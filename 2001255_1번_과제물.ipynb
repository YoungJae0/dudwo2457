```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,2*np.pi,100)
y=np.cos(x)
plt.plot(x,y)
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('Cos(x)')
plt.show
```
